```python
from newscatcher import Newscatcher
from newscatcher import describe_url
```

```python
website = 'https://valor.globo.com/financas/'
nc = Newscatcher(website = website, topic = 'business')


describe = describe_url(website)

print(describe['topics'])

results = nc.get_news()
articles = results['articles']
```
